# Data

```python
path_to_data1 = 'Dataset_small.raw.txt'
path_to_data2 = 'Dataset_big.raw.txt'
```

```python
# Read the contents of the file
def read_file_contents(data_path):
    with open(data_path, 'r') as file:
        contents = file.read()
    return contents

txt1 = read_file_contents(path_to_data1)
txt2 = read_file_contents(path_to_data2)
```

# Preprocess text and remove stop words


```python
stop_words = ['a', 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'for', 'from', 'has', 'he', 'in', 'is', 'it', 'its', 'of', 'on', 'that', 'the', 'to', 'was', 'were', 'will', 'with']
import re


def preprocess_text(corpus):
    # lowercase text
    corpus = corpus.lower()

    # remove numbers and punctuation
    corpus = re.sub(r'[<.,*?@:;=/\'(&)>]', '', corpus)
    corpus = re.sub(r'\b\d+\b', 'NUM', corpus)
    
    #remove stop words
    # Remove stop words and split into sentences
    results = []
    sentences = corpus.split('\n')  # Split text into sentences
    for sentence in sentences:
        words = sentence.split()  # Split sentence into words
        words = [word for word in words if word not in stop_words]  # Remove stop words
        processed_sentence = ' '.join(words)  # Join words back into a sentence
        results.append(processed_sentence)

    return results
```

```python
txt1 = preprocess_text(txt1)
txt2 = preprocess_text(txt2)
```

# Create word set by which we will have word vector

```python
words = []
for text in txt1:
    for word in text.split(' '):
        words.append(word)

txt1_words = set(words)
```

```python
words2 = []
for text in txt2:
    for word in text.split(' '):
        words2.append(word)

txt2_words = set(words2)
```

```python
#dictionary where each word in txt1_words is associated with a unique integer index.
word2int = {}

for i,word in enumerate(txt1_words):
    word2int[word] = i
```

```python
word2int2 = {}

for i,word in enumerate(txt2_words):
    word2int2[word] = i
```

# Data generation. We will use comtinious bag of words method  

```python
#hyperparameter word embedding size
```

```python
def get_windows(words, C):
    i = C
    while i < len(words) - C:
        center_word = words[i]
        context_words = words[(i - C):i] + words[(i+1):(i+C+1)]
        yield context_words, center_word
        i += 1
```

```python
data = []
for sentence in txt1:
    words = sentence.split()
    for context, center in get_windows(words,2):
        #print(context, center)
        data.append([context, center])
```

```python
data2 = []
for sentence in txt2:
    words = sentence.split()
    for context, center in get_windows(words,2):
        #print(context, center)
        data2.append([context, center])
```

```python
df = pd.DataFrame(data, columns = ['input', 'label'])
```

```python
df2 = pd.DataFrame(data2, columns = ['input', 'label'])
```

# One hot encoding and data for training a model

```python
# Create the word-to-index and index-to-word mappings
word2int = {word: idx for idx, word in enumerate(txt1_words)}
int2word = {idx: word for idx, word in enumerate(txt1_words)}

# Convert the data to numerical format
indexed_data = []
for input_seq, label in data:
    indexed_input = [word2int[word] for word in input_seq]
    indexed_label = word2int[label]
    indexed_data.append((indexed_input, indexed_label))


# Convert the data to one-hot vectors
def to_one_hot_encoding(data_point_index, vocab_size):
    one_hot_encoding = np.zeros(vocab_size)
    one_hot_encoding[data_point_index] = 1
    return one_hot_encoding

X = [] # input word
Y = [] # target word

for input_seq, label in indexed_data:
    X.append(to_one_hot_encoding(input_seq, len(txt1_words)))
    Y.append(to_one_hot_encoding(label, len(txt1_words)))

# Convert them to torch tensors
X_train = torch.tensor(X, dtype=torch.float32)
Y_train = torch.tensor(Y, dtype=torch.float32)
```

```python
# Create the word-to-index and index-to-word mappings
word2int2 = {word: idx for idx, word in enumerate(txt2_words)}
int2word2 = {idx: word for idx, word in enumerate(txt2_words)}

# Convert the data to numerical format
indexed_data2 = []
for input_seq, label in data2:
    indexed_input = [word2int2[word] for word in input_seq]
    indexed_label = word2int2[label]
    indexed_data2.append((indexed_input, indexed_label))


# Convert the data to one-hot vectors
def to_one_hot_encoding(data_point_index, vocab_size):
    one_hot_encoding = np.zeros(vocab_size)
    one_hot_encoding[data_point_index] = 1
    return one_hot_encoding

X2 = [] # input word
Y2 = [] # target word

for input_seq, label in indexed_data2:
    X2.append(to_one_hot_encoding(input_seq, len(txt2_words)))
    Y2.append(to_one_hot_encoding(label, len(txt2_words)))

# Convert them to torch tensors
X_train2 = torch.tensor(X2, dtype=torch.float32)
Y_train2 = torch.tensor(Y2, dtype=torch.float32)
```

# Word2vec model. Dataset 1.

```python
import torch.nn as nn

# Define the Word2Vec model
class Word2Vec(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Word2Vec, self).__init__()
        self.hidden_layer = nn.Linear(input_dim, output_dim)
        self.output_layer = nn.Linear(output_dim, input_dim)
    
    def forward(self, x):
        hidden = self.hidden_layer(x)
        embedded = self.output_layer(hidden)
        return embedded
```

# Training the data

```python
import torch.optim as optim

# Define hyperparameters
EMBEDDING_DIM = 2
lr = 0.05
num_epochs = 50

ONE_HOT_DIM = len(txt1_words)

# Create the Word2Vec model
model = Word2Vec(ONE_HOT_DIM, EMBEDDING_DIM)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=lr)

# Training loop
for epoch in range(num_epochs):
    # Forward pass
    output = model(X_train)
    
    # Compute loss
    loss = criterion(output, X_train)
    
    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # Print the loss
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

```

# Word2vec model. Dataset 2.

```python
class Word2Vec2(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Word2Vec, self).__init__()
        self.hidden_layer1 = nn.Linear(input_dim, hidden_dim)
        self.hidden_layer2 = nn.Linear(hidden_dim, output_dim)
        self.output_layer = nn.Linear(output_dim, input_dim)
    
    def forward(self, x):
        hidden1 = self.hidden_layer1(x)
        hidden2 = self.hidden_layer2(hidden1)
        embedded = self.output_layer(hidden2)
        return embedded

```

```python
# Define hyperparameters
EMBEDDING_DIM2 = 2
lr = 0.001
num_epochs = 20

#input_dim = 100  # Dimensionality of input
#output_dim = 50  # Dimensionality of output/embedding
num_hidden_layers = 2  # Number of hidden layers
hidden_units = 64  # Number of hidden units in each hidden layer


input_dim = len(txt2_words)

# Create the Word2Vec model
model2 = Word2Vec(input_dim, EMBEDDING_DIM2)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model2.parameters(), lr=lr)

# Training loop
for epoch in range(num_epochs):
    # Forward pass
    output = model2(X_train2)
    
    # Compute loss
    loss = criterion(output, X_train2)
    
    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # Print the loss
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

```

# Word clusters

```python
from sklearn.cluster import KMeans


# Apply k-Means clustering
num_clusters = 6  # Define the number of clusters
n_init = 10  # Define the number of times the k-means algorithm will be run

kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=n_init)
kmeans.fit(df[['Embedding 1', 'Embedding 2']])

# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Print the word clusters
for cluster in range(num_clusters):
    words_in_cluster = df[df['Cluster'] == cluster]['word'].tolist()
    print(f"Cluster {cluster}: {words_in_cluster}")
```
